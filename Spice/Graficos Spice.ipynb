```python
import ltspice
import numpy as np
import matplotlib.pyplot as plt
import os

# ----------------------------     Graficos de los circuitos de Spice     --------------------------------------


#------------Load the signals from the .raw file
raw_file = 'Circuito_Diodo\\Diodo.raw'
l = ltspice.Ltspice(raw_file) 
# Make sure that the .raw file is located in the correct path
l.parse()
#-----------------------------------------------


#----------------------------Obtain basic signals         
VD = l.get_data('V(VD)')
ID = l.get_data('I(D1)')
ID = (ID) * 1E+3
#------------------------------------------------

chargeTime = np.linspace(0, 250, 1000)


#Set mplt and min Voltages and currents
V_upper_limit = 0.6
V_lower_limit = -0.5
I_upper_limit = 1
I_lower_limit = -0.05

fig = plt.figure(figsize=(7, 5))
fig.subplots_adjust(hspace=0.5)

#Current plot
plt.title("$I_D \ vs \ V_D$")       #title
plt.xlabel("$V_D \ [V]$")
plt.ylabel("$I_D \ [mA]$")
plt.xlim(V_lower_limit, V_upper_limit)
plt.ylim(I_lower_limit, I_upper_limit)
plt.grid()

plt.plot(VD , ID, 'b-')

#plt.savefig('Grafico Circuito Diodo.pdf', format='pdf')
plt.show()
```

```python
import ltspice
import numpy as np
import matplotlib.pyplot as plt
import os

# ----------------------------     Grafico BJT Curva de Entrada     --------------------------------------


#------------Load the signals from the .raw file
raw_file = 'Circuito_BJT\\Circuito_BJT_Curva_Entrada.raw'
l = ltspice.Ltspice(raw_file) 
# Make sure that the .raw file is located in the correct path
l.parse()
#-----------------------------------------------


#----------------------------Obtain basic signals         
VBE = l.get_data('V(vbe)')
IB = l.get_data('I(Rb)')
IB = (-IB) * 1E+3
#------------------------------------------------

chargeTime = np.linspace(0, 250, 1000)


#Set mplt and min Voltages and currents
V_upper_limit = 0.9
V_lower_limit = -0.5
I_upper_limit = 45
I_lower_limit = -1

fig = plt.figure(figsize=(7, 5))
fig.subplots_adjust(hspace=0.5)

#Current plot
plt.title("Curva de Entrada del BJT")       #title
plt.xlabel("$V_{BE}\ [V]$")
plt.ylabel("$I_B \ [mA]$")
plt.xlim(V_lower_limit, V_upper_limit)
plt.ylim(I_lower_limit, I_upper_limit)
plt.grid()

plt.plot(VBE , IB, 'b-')

#plt.savefig('Grafico Curva de Entrada BJT', format='pdf')
plt.show()
```
