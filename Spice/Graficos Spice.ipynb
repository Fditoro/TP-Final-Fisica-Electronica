```python
import ltspice
import numpy as np
import matplotlib.pyplot as plt
import os

# ----------------------------     Graficos de los circuitos de Spice     --------------------------------------

VDQ = 0.510417
IDQ = 0.00029454E+3

#------------Load the signals from the .raw file
raw_file = 'Circuito_Diodo\\Diodo.raw'
l = ltspice.Ltspice(raw_file) 
# Make sure that the .raw file is located in the correct path
l.parse()
#-----------------------------------------------


#----------------------------Obtain basic signals         
VD = l.get_data('V(VD)')
ID = l.get_data('I(D1)')
ID = (ID) * 1E+3
#------------------------------------------------


#Set mplt and min Voltages and currents
V_upper_limit = 0.58
V_lower_limit = -0.15
I_upper_limit = 0.85
I_lower_limit = -0.05

fig = plt.figure(figsize=(6, 4))

#Current plot
plt.xlabel("$V_D \ [V]$")
plt.ylabel("$I_D \ [mA]$")

#---------      Anotaciones     --------------
plt.plot([VDQ,VDQ], [IDQ, I_lower_limit], linestyle = '--', linewidth = 1.5,  color = 'black')
plt.plot([V_lower_limit,VDQ], [IDQ, IDQ], linestyle = '--', linewidth = 1.5,  color = 'black')
plt.scatter(VDQ, IDQ, color='red', s = 30)
plt.scatter(VDQ, I_lower_limit, color='black', s = 1, label = '$V_{DQ} = 0,510 \ V$')
plt.scatter(V_lower_limit, IDQ, color='black', s = 1, label = '$I_{DQ} = 294,54 \ uA$')
plt.annotate(r"$Q$", xy=(VDQ, IDQ), xytext=(-7, 7),textcoords='offset points', ha='center', fontsize = 15, color = 'red')

plt.xlim(V_lower_limit, V_upper_limit)
plt.ylim(I_lower_limit, I_upper_limit)
plt.grid()

plt.plot(VD , ID, 'b-')
plt.legend(fontsize = 16)

#plt.savefig('Grafico Circuito Diodo.pdf', format='pdf')
plt.show()
```
