```python
import ltspice
import numpy as np
import matplotlib.pyplot as plt
import os

# ----------------------------     Graficos de los circuitos de Spice     --------------------------------------

VDQ = 0.10916
IDQ = 334.0675618E-6

#------------Load the signals from the .raw file
raw_file = 'Circuito_Diodo\\Diodo.raw'
l = ltspice.Ltspice(raw_file) 
# Make sure that the .raw file is located in the correct path
l.parse()
#-----------------------------------------------


#----------------------------Obtain basic signals         
VD = l.get_data('V(VD)')
ID = l.get_data('I(D1)')
ID = (ID) * 1E+3
#------------------------------------------------


#Set mplt and min Voltages and currents
V_upper_limit = 0.58
V_lower_limit = -0.15
I_upper_limit = 0.85
I_lower_limit = -0.05

fig = plt.figure(figsize=(6, 7))
fig.subplots_adjust(hspace=0.5)

#Current plot
plt.title("$I_D \ vs \ V_D$")       #title
plt.xlabel("$V_D \ [V]$")
plt.ylabel("$I_D \ [mA]$")

#---------      Anotaciones     --------------
plt.plot([VDQ,VDQ], [IDQ, I_lower_limit], linestyle = '--', linewidth = 1.5,  color = 'black')
plt.scatter(VDQ, IDQ, color='red', s = 30)
plt.scatter(VDQ, I_lower_limit, color='black', s = 1, label = '$V_{DQ} = 0,109 \ V$')
plt.scatter(V_lower_limit, IDQ, color='black', s = 1, label = '$I_{DQ} = 334,068 \ uA$')
plt.annotate(r"$Q$", xy=(VDQ, IDQ), xytext=(5, 5),textcoords='offset points', ha='center', fontsize = 15, color = 'red')

plt.xlim(V_lower_limit, V_upper_limit)
plt.ylim(I_lower_limit, I_upper_limit)
plt.grid()

plt.plot(VD , ID, 'b-')
plt.legend(fontsize = 16)

plt.savefig('Grafico Circuito Diodo.jpg', format='jpg')
plt.show()
```
